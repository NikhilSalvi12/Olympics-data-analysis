### Most succesful Athletes (top 10): summer olympics

```python
top_athletes = pd.DataFrame(summer.Athlete.value_counts().head(10)).reset_index()
top_athletes.columns=["Athlete","Medal"]
top_athletes
```

### Vizualizing Succesful countries (top 10)

```python
plt.figure(figsize=(15,10))
ax = sns.countplot(y="Country",data=summer,hue="Medal",order=summer.Country.value_counts().iloc[:10].index)

for p in ax.patches:
    ax.annotate('{:}'.format(p.get_x()),xy = (p.get_height(),p.get_x()+1))
plt.show()
```
